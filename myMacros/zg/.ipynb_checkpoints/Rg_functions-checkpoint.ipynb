# Development of functions related to Rg

In this analysis, I use Rg = sqrt(Deta^2 + Dphi^2), Deta, Dphi the angular distance between the two sub-jets.

## Imports

## Examine files

```c++
string path_incl = "/data_CMS/cms/mnguyen//bJet2022/qcdMC/SD/merged_HiForestAOD.root";
string path_bJet = "/data_CMS/cms/mnguyen//bJet2022/bJetMC/SD/merged_HiForestAOD.root";
TFile *f;
TTree *t;
TTree *HiTree;
```

```c++
f = new TFile(path_bJet.c_str());
//f->ls();
```

```c++
t = (TTree *) f->Get("ak4PFJetAnalyzer/t");
HiTree = (TTree *) f->Get("hiEvtAnalyzer/HiTree");
//t->Print();
```

```c++
Long64_t nentries = t->GetEntries();
```

```c++
Int_t ngen;
t->SetBranchAddress("ngen", &ngen);

Float_t psjt1Eta[30];
t->SetBranchAddress("psjt1Eta", psjt1Eta);

Float_t psjt2Eta[30];
t->SetBranchAddress("psjt2Eta", psjt2Eta);

Float_t psjt1Phi[30];
t->SetBranchAddress("psjt1Phi", psjt1Phi);

Float_t psjt2Phi[30];
t->SetBranchAddress("psjt2Phi", psjt2Phi);
```

```c++
Float_t parpt[30];
t->SetBranchAddress("parpt", parpt);
```

```c++
Float_t psjt2Pt[30];
t->SetBranchAddress("psjt2Pt", psjt2Pt);
```

```c++
Float_t weight;
HiTree->SetBranchAddress("weight", &weight);
```

```c++
Int_t npar;
t->SetBranchAddress("npar", &npar);
```

```c++
t->SetBranchStatus("*", 0);
t->SetBranchStatus("ngen", 1);
t->SetBranchStatus("nref", 1);
t->SetBranchStatus("psjt1Pt", 1);
t->SetBranchStatus("psjt2Pt", 1);
t->SetBranchStatus("jtpt", 1);
t->SetBranchStatus("genpt", 1);
t->SetBranchStatus("psjt1Eta", 1);
t->SetBranchStatus("psjt2Eta", 1);
t->SetBranchStatus("psjt1Phi", 1);
t->SetBranchStatus("psjt2Phi", 1);
t->SetBranchStatus("parpt", 1);
t->SetBranchStatus("npar", 1);

HiTree->SetBranchStatus("*", 0);
HiTree->SetBranchStatus("weight", 1);
```

```c++
hnew = new TH2F("hnew", "", 20, 0.01, 1.5, 16, 20, 100);
counter = 0;
newcounter = 0;
std::cout << "started" << endl;
for (Long64_t i = 0; i < nentries; i++) {
    
    t->GetEntry(i);
    HiTree->GetEntry(i);
    if (i%1000000 == 0) {std::cout << "i = " << i << endl;}
    for (Int_t j = 0; j < ngen; j++) {
        Float_t rg = 0;
        if (psjt2Pt[j] > 0) {
            Float_t Deta = psjt1Eta[j] - psjt2Eta[j];
            Float_t Dphi = acos(cos(psjt1Phi[j] - psjt2Phi[j]));
            rg = sqrt(Dphi*Dphi+Deta*Deta);
            hnew->Fill(rg, parpt[j], weight);
        } 
        hnew->Fill(rg, parpt[j], weight);
    }
}
hnew->SetXTitle("rg");
hnew->SetYTitle("parpt");
hnew->SetStats(0);
hnew->Draw("COLZ");
c->Draw();
```
